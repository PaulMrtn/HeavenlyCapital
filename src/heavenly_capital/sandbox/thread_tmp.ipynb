```python
from heavenly_capital.core.runtime_config import ThreadConfig
from heavenly_capital.core.thread_manager import ThreadManager
import time
from threading import current_thread


def test_task(name, duration=0.5):
    """Tâche de test simple"""
    print(f"{current_thread().name} commence {name}")
    time.sleep(duration)
    print(f"{current_thread().name} termine {name}")

if __name__ == "__main__":
    # Initialisation
    manager = ThreadManager()
    manager.initialize_pools()

    # Affichage des pools
    print("Pools disponibles :", list(manager.pools.keys()))

    # Soumettre des tâches de test
    for i in range(5):
        manager.submit("STANDARD", test_task, f"Tâche-{i}", 0.2)

    # Explorer l'état des pools et workers
    for pool_name, pool in manager.pools.items():
        print(f"\nPool {pool_name}: queue_size={pool.queue_size}")
        for w in pool.workers:
            print(f"  Worker {w.name}: alive={w.is_alive()}, last_error={w.last_error}")

    # Attendre que toutes les tâches soient terminées
    time.sleep(1)

    # Vérifier l'état à nouveau
    for pool_name, pool in manager.pools.items():
        print(f"\nAprès traitement - Pool {pool_name}: queue_size={pool.queue_size}")
        for w in pool.workers:
            print(f"  Worker {w.name}: alive={w.is_alive()}, last_error={w.last_error}")

    # Arrêt propre
    manager.shutdown()
    print("\nTous les pools ont été arrêtés")
```

```python
import time
from threading import current_thread

# Adapte ces imports selon ton projet
from heavenly_capital.core.thread_manager import get_thread_manager  # ou: ThreadManager
from heavenly_capital.core.runtime_config import ThreadConfig
from heavenly_capital.core.system_manager import SystemPorts


def test_task(name: str, duration: float = 0.2) -> None:
    print(f"[{current_thread().name}] start {name}")
    time.sleep(duration)
    print(f"[{current_thread().name}] end   {name}")


manager = get_thread_manager()  # singleton du projet
manager.configure(config=ThreadConfig(), ports=None)
manager.start()

# Choisis un pool existant dans ta config (ex: "CRITICAL" ou "STANDARD")
pool_name = "CRITICAL"

for i in range(10):
    manager.submit(pool_name, test_task, f"job-{i}", 0.1)

# Petit état
pool = manager.get_pool(pool_name)
print(f"queue_size={pool.queue_size}")
for w in pool.workers:
    print(f"worker={w.name} alive={w.is_alive()} last_error={w.last_error}")

```
